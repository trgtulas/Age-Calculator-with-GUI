```python
from tkinter import *
from tkinter import messagebox
import datetime
import time
import os
import googletrans


now = datetime.datetime.now()

hesaplama = Tk()
hesaplama.title("Age Calculator")

Canvas = Canvas(hesaplama, width= 500, height= 300)
Canvas.pack()

yas_girdi = Frame(Canvas, bg = "#B5A300")
yas_girdi.place( relx= 0.1, rely= 0.1, relwidth= 0.8,
                relheight= 0.30)

sonuc_girdi = Frame(Canvas, bg = "#B5A300")
sonuc_girdi.place(relwidth= 0.8, relheight= 0.48, relx= 0.1,
                  rely= 0.42)

results_age = Label(sonuc_girdi, text ="Yaşınız:",
                bg = "#B5A300", 
                font = "Verdana 12 bold")
results_age2 = Label(sonuc_girdi, text = "Doğduğunuz gün:", 
                bg = "#B5A300", font= "Verdana 12 bold")
results_age3 = Label(sonuc_girdi, text = "Mevsim: ", 
                bg = "#B5A300", font= "Verdana 12 bold")

results_age.pack(anchor=NW, padx = 10, pady = 10)
results_age2.pack(anchor = NW, padx = 10, pady= 10)
results_age3.pack(side = LEFT, padx = 10, pady = 10)
    


gun_girdi = Label(yas_girdi,text="Gün", bg = "#B5A300",
                  font= "Verdana 12 bold")
gun_girdi.place(relx = 0.01, rely = 0.22)

gun_girdi_opsiyon = IntVar(yas_girdi)
gun_girdi_opsiyon.set("1")

list1 = list(range(1,32))
gun_girdi_acilir_menu = OptionMenu(yas_girdi, gun_girdi_opsiyon,
                                   *list1)
gun_girdi_acilir_menu.place(relx = 0.12, rely = 0.15 )

ay_girdi = Label(yas_girdi, text="Ay", bg = "#B5A300",
                 font= "Verdana 12 bold")
ay_girdi.place(relx = 0.33, rely = 0.22)

list2 = ("Ocak","Şubat","Mart","Nisan","Mayıs","Haziran",
         "Temmuz","Ağustos","Eylül", "Ekim", "Kasım", "Aralık")
ay_girdi_opsiyon = StringVar(yas_girdi)
ay_girdi_opsiyon.set("Ocak")
ay_girdi_acilir_menu = OptionMenu(yas_girdi, ay_girdi_opsiyon,
                                  *list2)
ay_girdi_acilir_menu.place(relx = 0.42, rely = 0.15 )
yıl_girdi = Label(yas_girdi, text="Yıl", bg = "#B5A300",
                  font= "Verdana 12 bold")
yıl_girdi.place(relx = 0.67, rely = 0.22)

yıl_listesi = list(range(1960,now.year+1))
yıl_girdi_opsiyon = IntVar(yas_girdi)
yıl_girdi_opsiyon.set(1999)
yıl_girdi_acilir_menu = OptionMenu(yas_girdi, yıl_girdi_opsiyon,
                                   *yıl_listesi)
yıl_girdi_acilir_menu.place(relx = 0.75, rely = 0.15 )



def hesapla():
    gun = gun_girdi_opsiyon.get()
    ay = ay_girdi_opsiyon.get()
    yil = yıl_girdi_opsiyon.get()
    
    gun = int(gun)
    yil = int(yil)
    
    day = now.day-gun
    month = now.month - (list2.index(ay) + 1)
    year = now.year-yil
    
    if day < 0:
        day += 30
        month -= 1
    else:
        year = now.year - yil
        month = now.month - (list2.index(ay) + 1)
        day = now.day-gun
    if month < 0:
        month += 12
        year -= 1
    else:
        year = now.year - yil
        month = now.month - (list2.index(ay) + 1)
        day = now.day-gun
    
    if ((ay == "Aralık") or (ay == "Ocak") or (ay == "Şubat")):
        season = "Kış"
    elif ((ay == "Mart") or (ay == "Nisan") or (ay == "Mayıs")):
        season = "İlkbahar"
    elif ((ay == "Haziran") or (ay == "Temmuz") or (ay == "Ağustos")):
        season = "Yaz"
    else:
        season = "Sonbahar"
        
        
    messagebox_results = "Completed"
    messagebox.showinfo("Completed", messagebox_results)
    
    birthday1 = datetime.date(yil, list2.index(ay) + 1, gun).strftime("%A")
    
    if (birthday1 == "Monday"):
        birthday = "Pazartesi"
    elif (birthday1 == "Tuesday"):
        birthday = "Salı"
    elif (birthday1 == "Wednesday"):
        birthday = "Çarşamba"
    elif (birthday1 == "Thursday"):
        birthday = "Perşembe"
    elif (birthday1 == "Friday"):
        birthday = "Cuma"
    elif (birthday1 == "Saturday"):
        birthday = "Cumartesi"
    else:
        birthday = "Pazar"
    
    results_age.config(text = "Yaşınız: {} yıl, {} ay, {} gün".format(year, month, day))
    results_age2.config(text = "Doğduğunuz gün: {}".format(birthday))
    results_age3.config(text = "Mevsim: {}".format(season))
    
buton = Button(yas_girdi, text="Hesapla",
               bg = "#B5A300", font="Verdana 10 bold",
              command =hesapla)
buton.pack(side=BOTTOM, pady = 7)


hesaplama.mainloop()
```
